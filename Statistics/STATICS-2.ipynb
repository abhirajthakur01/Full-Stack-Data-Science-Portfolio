# zscore

```python
mean = df["CoursePrice"].mean()
std = df["CoursePrice"].std()
mean,std
```

```python
df["z_score"] = (df["CoursePrice"]-mean)/std
```

```python
df[["CoursePrice","z_score"]]
```

```python
sns.histplot(df["z_score"],kde = True)
plt.plot()
```

#### percentage of students who paid between 73 and 100 in course price?

```python
x = (73-mean)/std
y = (100 - mean)/std
x,y
```

```python
import scipy.stats as stats
```

```python
lower = stats.norm.cdf(x)
upper = stats.norm.cdf(y)
```

```python
lower,upper
```

```python
percentage = (upper-lower)*100
```

```python
remaining_percentage = 100 - percentage
```

```python
percentage,remaining_percentage
```
