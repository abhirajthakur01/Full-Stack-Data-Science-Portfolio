#### FIVE NO SUMMARY

```python
df['CoursePrice']
```

```python
df['CoursePrice']
```

```python
Q1  = np.percentile(df['CoursePrice'],25)
Q3  = np.percentile(df['CoursePrice'],75)
IQR = Q3-Q1
LF  = np.round(Q1-1.5*IQR,2)
UF  = np.round(Q3 +1.5*IQR,2)
```

```python
print(f"upper fence and lower fence are :{UF} and {LF}")
```

```python
plt.figure(figsize = (10,3))
sns.boxplot(x = df['CoursePrice'])
plt.xticks([LF,df['CoursePrice'].min(),Q1,Q3,df['CoursePrice'].max(),UF])
plt.show()
```

###### Z-score

##### Z-score = X- mean/std

```python
mean = df['CoursePrice'].mean()
std = df['CoursePrice'].std()
mean,std
```

```python
df['Z_score_CP'] = (df['CoursePrice']-mean)/std
```

```python
df[['CoursePrice','Z_score_CP']]
```

```python
df['Z_score_CP']
```

```python
sns.histplot(df['Z_score_CP'],kde=True) #kde = kernel density estimator
plt.show()
```

###### Percentage of the student who have paid course price btw 70 and 100 ?

```python
z_lower = (70-mean)/std
z_upper = (100-mean)/std
z_lower,z_upper
```

```python
import scipy.stats as stats
```

```python
lower  = stats.norm.cdf(z_lower)
upper  = stats.norm.cdf(z_upper)
```

```python
percentage  = (upper-lower)*100
percentage
```
