```python
x1 = np.random.normal(0,4,1000)
x2 = np.random.normal(-2,2,1000)
x3 = np.random.normal(1,5,1000)
```

```python
plt.hist(x1)
plt.hist(x2)
plt.hist(x3)
```

```python
plt.hist(x1 , alpha = 0.5, density = True, bins = 40)
plt.hist(x2, alpha = 0.5, density = True, bins = 40)
plt.hist(x3, alpha = 0.5, density = True, bins = 40)
```
