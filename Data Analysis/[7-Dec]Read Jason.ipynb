# Pandas have a nice inbuilt fuction called json_normalize() to flatten the simple to moderately semi-structured nested JSON structures to flat tables.

```python
df = pd.DataFrame({
    'student_id': [1, 2, 3],
    'student_name': ['Alice', 'Bob', 'Chris'],
    'student_info': [
        {'gender': 'F', 'age': 20},
        {'gender': 'M', 'age': 22},
        {'gender': 'M', 'age': 33}
    ]
})
```

```python
df2=pd.json_normalize(df['student_info'])
```

```python
df2
```

```python
df3=pd.concat([df,df2],axis=1)
```

```python
df3
```

```python
df3.drop(columns='student_info',inplace=True)
```
