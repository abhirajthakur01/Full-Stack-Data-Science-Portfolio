# unstacked and stacked bar plots

```python
plotdata = pd.DataFrame({
    "pies_2018":[40, 12, 10, 26, 36],
    "pies_2019":[19, 8, 30, 21, 38],
    "pies_2020":[10, 10, 42, 17, 37]
    }, 
    index=["Dad", "Mam", "Bro", "Sis", "Me"]
)
plotdata.head()
```

```python
plotdata.plot(kind='bar',figsize=(10,4))
```

```python
import matplotlib.pyplot as plt
```

```python
plotdata.plot(kind='bar',figsize=(15,6))
plt.title('mince pie consumption study',fontdict={'fontsize':32})
plt.xlabel('family members',fontdict={'fontsize':20})
plt.ylabel('pies consumption',fontdict={'fontsize':20})
```

```python
plotdata.plot(kind='bar',figsize=(15,6),stacked=True)
plt.title('mince pie consumption study',fontdict={'fontsize':32})
plt.xlabel('family members',fontdict={'fontsize':20})
plt.ylabel('pies consumption',fontdict={'fontsize':20})
plt.legend(['pies_2018','pies_2019','pies_2020'],loc='center left',fontsize=20)
```

```python
plotdata.plot(kind='bar',figsize=(15,6),stacked=True)
plt.title('mince pie consumption study',fontdict={'fontsize':32})
plt.xlabel('family members',fontdict={'fontsize':20})
plt.ylabel('pies consumption',fontdict={'fontsize':20})
plt.legend(['pies_2018','pies_2019','pies_2020'],loc='upper left',fontsize=20)
```

```python
plotdata.plot(kind='bar',figsize=(15,6),stacked=True)
plt.title('mince pie consumption study',fontdict={'fontsize':32})
plt.xlabel('family members',fontdict={'fontsize':20})
plt.ylabel('pies consumption',fontdict={'fontsize':20})
plt.legend(['pies_2018','pies_2019','pies_2020'],loc='upper right',fontsize=20)
```

```python
plotdata[["pies_2018","pies_2020","pies_2019"]].plot(kind='bar',stacked=True)
plt.title('Mince Pie Consumption Study',fontdict={'fontsize':32})
plt.xlabel('Family Members',fontdict={'fontsize':20})
plt.ylabel('Pies Consumption',fontdict={'fontsize':20})
```

```python
plotdata[["pies_2020","pies_2018","pies_2019"]].plot(kind='bar',figsize=(15,6),stacked=True).legend(ncol=3,loc='center left',title='year of eating')
plt.title('Mince Pie Consumption Study',fontdict={'fontsize':32})
plt.xlabel('Family Members',fontdict={'fontsize':20})
plt.ylabel('Pies Consumption',fontdict={'fontsize':20})
```
