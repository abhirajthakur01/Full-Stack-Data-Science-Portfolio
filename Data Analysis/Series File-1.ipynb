```python
# Creating the Series 
sr = pd.Series([10, 25, 3, 25, 24, 6]) 
  
# Create the Index 
index_ = ['Coca Cola', 'Sprite', 'Coke', 'Fanta', 'Dew', 'ThumbsUp'] 
  
# set the index 
sr.index = index_ 
  
# Print the series 
print(sr) 

# return the mean 
result = sr.mean() 
  
# Print the result 
print(result) 
```

```python
#creating the series 
sr = pd.Series([19.5, 16.8, None, 22.78, 16.8, 20.124, None, 18.1002, 19.5])

#print the series 
print(sr)

#return the mean 
#dont skip all the missing values
result1 = sr.mean(skipna=False)

#print the result
print(result1)

#skip all the missing values
result = sr.mean(skipna = True)

#print the result 
print(result)
```
