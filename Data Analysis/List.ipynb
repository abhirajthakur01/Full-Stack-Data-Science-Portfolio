```python
#Adding and updating lists
courses=['OOP','Networking','MIS','Project']
students=['rahul','rohit','sagar','mohit','kajal']
OOP_marks = [65,85,95]
print(OOP_marks)
```

```python
#updating new elements
OOP_marks[0] = 70
OOP_marks[1] = 45
print(OOP_marks)
```

```python
#extend list with another list
list1 = [88,93]
OOP_marks.extend(list1)
print(OOP_marks)
```

```python
#Deleting an element from a list 
OOP_marks = [70, 45, 92, 50, 77, 45]
print(OOP_marks)

#Delete an element using del
del OOP_marks[0]
print(OOP_marks)

#Remove an element using remove()
OOP_marks.remove(45)
print(OOP_marks)

#Remove an element using pop()
OOP_marks.pop()
print(OOP_marks)
```

```python
#find the list length
OOP_marks = [70, 45, 92, 50, 77, 45]
CSO_marks = [80,59,76]
print("Length of lists")
print(len(OOP_marks))
print(len(CSO_marks))


#concatinate lists
print("Concatenation of lists")
print(OOP_marks+CSO_marks)

#Repeat an element in a list
print('Repeat an elements in a lists')
print(['code']*5)

#check elements is in list or not
print(70 in OOP_marks)

#Access elements using for loop
for i in OOP_marks:
    print(i, end='*')
      
```
