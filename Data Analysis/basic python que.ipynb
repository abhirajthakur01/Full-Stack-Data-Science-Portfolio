```python
def order_food(*args,**kwargs):
    print("order summary")
    print("food items ordered:")
    for i in args:
        print(f"items:{i}")
    if kwargs:
        print("\n additional details")
        for key,value in kwargs.items():
            print(f"{key}:{value}")
```
