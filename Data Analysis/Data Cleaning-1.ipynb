```python
#case2
df = pd.read_csv("mpg-data.csv")
```

```python
df.isna()
```

```python
df.isna().sum().to_frame()
```

```python
df1 = df.isna().sum().to_frame()
```

```python
df1
```

```python
df1.rename(columns={0:'Null'},inplace=True)
```

```python
df1
```

```python
df2 = df.notna().sum().to_frame()
df2.rename(columns={0:'Not Null'},inplace=True)
```

```python
print(df2)
```

```python
df3 = pd.concat([df1,df2],axis=1)
df3.reset_index(inplace=True)
```

```python
df3
```

```python
df3.rename(columns={'index':'Category'})
```

```python
df3
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(df3['index'],df3['Null'] )
plt.xticks(rotation=45)
plt.xlabel('category')
plt.ylabel('No.of Null values')
plt.title('No. of Null values against category')
```
