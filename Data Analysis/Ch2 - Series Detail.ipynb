# Create a series

### Creating an empty series

Series() function of Pandas is used to create a series. A basic series, which can be created is an Empty Series. 

```python
S1 = pd.Series()
print(S1)
```

### Creating a series from array

In order to create a series from NumPy array, we have to import numpy module and have to use array() function. 

```python
# Simple array
d1 = np.array(['c','o','d','i','n','g'])
S2 = pd.Series(d1)
print(S2)
```

### Creating a series from array with an index

In order to create a series by explicitly proving index instead of the default, we have to provide a list of elements to the index parameter with the same number of elements as it is an array. 

```python
data = np.array(['c','o','d','i','n','g'])
S3 = pd.Series(data,index=[10,11,12,13,14,15])
print(S3)
```

### Creating a series from Lists

In order to create a series from list, we have to first create a list after that we can create a series from list. 

```python
L1 = ['c','o','d','i','n','g']
S4 = pd.Series(L1)
print(S4)
```

### Creating a series from Dictionary

In order to create a series from the dictionary, we have to first create a dictionary after that we can make a series using dictionary. Dictionary keys are used to construct indexes of Series.  

```python
# a simple dictionary
dict = {'K1': 10,
        'K2': 20,
        'K3': 30}
 
# create series from dictionary
ser = pd.Series(dict)
 
print(ser)
```

```python
dict = {'K1':10,'K2':20,'K3':30}
ser = pd.Series(dict, index = ['a','b','c','K1','K2'])
ser
```

### Creating a series from Scalar value

In order to create a series from scalar value, an index must be provided. The scalar value will be repeated to match the length of the index. 

```python
# Giving a scalar value with index
S5 = pd.Series(10, index = [0,1,2,3,4,5])
print(S5)
```

```python
# Giving a scalar value with index
S5 = pd.Series(10, index = range(1,20,2))
print(S5)
```

```python
# Giving a scalar value with index
S5 = pd.Series(10)
print(S5)
```

#### Pandas Series.mean(), min(), max(), sum(), std()

Pandas Series.mean() function return the mean of the underlying data in the given Series object.

```python
# Creating the Series 
sr = pd.Series([10, 25, 3, 25, 24, 6]) 
  
# Create the Index 
index_ = ['Coca Cola', 'Sprite', 'Coke', 'Fanta', 'Dew', 'ThumbsUp'] 
  
# set the index 
sr.index = index_ 
  
# Print the series 
print(sr) 

# return the mean 
result = sr.mean() 
  
# Print the result 
print(result) 
```

```python
# Creating the Series 
sr = pd.Series([19.5, 16.8, None, 22.78, 16.8, 20.124, None, 18.1002, 19.5]) 
  
# Print the series 
print(sr) 

# return the mean 
# dont skip all the missing values 
result1 = sr.mean(skipna=False) 
  
# Print the result 
print(result1) 

# skip all the missing values 
result = sr.mean(skipna = True) 
  
# Print the result 
print(result) 
```

```python
# Creating the Series 
sr = pd.Series([10, 25, 3, 25, 24, 6]) 
  
# Create the Index 
index_ = ['Coca Cola', 'Sprite', 'Coke', 'Fanta', 'Dew', 'ThumbsUp'] 
  
# set the index 
sr.index = index_ 
  
# Print the series 
print(sr) 

print("Maximum Value:- ",sr.max()) 
print("Mimimum Value:- ",sr.min()) 
print("Sum of Value:- ",sr.sum()) 
print("Standard Deviation Value:- ",sr.std()) 
```

#### Pandas Series.isna() , Pandas Series.notna(), Pandas Series.dropna()

Pandas Series.isna() function detect missing values in the given series object. It return a boolean same-sized object indicating if the values are NA. Missing values gets mapped to True and non-missing value gets mapped to False. Pandas Series.notna() function Detect existing (non-missing) values. This function return a boolean object having the size same as the object, indicating if the values are missing values or not. Non-missing values get mapped to True.
Pandas Series.dropna() function return a new Series with missing values in the given series object removed.

```python
import numpy as np
```

```python
# Creating the Series 
sr = pd.Series([10, 25, 3, np.nan,25, 24, np.nan]) 
  
# Print the series 
print(sr) 
print(sr.isna())
print(sr.isnull())
print(sr.isna().sum())
print(sr.isnull().sum())
print(sr.notna())
print(sr.notnull())
```

```python
# Creating the Series 
sr = pd.Series([10, np.nan, 3,np.nan, 24, None]) 
  
# Print the series 
print(sr) 
print(sr.isna())
print(sr.isnull())
print("Total Missing Values:- ",sr.isna().sum())
```

```python
# Creating the Series 
sr = pd.Series([10, np.nan, 3,np.nan, 24, None]) 
  
# Print the series 
print(sr) 
print(sr.notna())
print(sr.notnull())
print("Total Not Missing Values:- ",sr.notnull().sum())
```

#### Pandas Series.drop()

Pandas Series.drop() function return Series with specified index labels removed. It remove elements of a Series based on specifying the index labels.

```python
# Creating the Series 
sr = pd.Series([80, 25, 3, 25, 24, 6]) 
print(sr)
  
# Create the Index 
index_ = ['Coca Cola', 'Sprite', 'Coke', 'Fanta', 'Dew', 'ThumbsUp'] 
  
# set the index 
sr.index = index_ 
  
# Print the series 
print(sr) 
```

```python
# Creating the Series 
sr = pd.Series([80, 25, 3, 25, 24, 6]) 
print(sr)
# Create the Index 
index_ = ['Coca Cola', 'Sprite', 'Coke', 'Fanta', 'Dew', 'ThumbsUp'] 
  
# set the index 
sr.index = index_ 
  
# Print the series 
print(sr) 

# drop the passed labels 
result = sr.drop(labels = ['Sprite', 'Dew'])  
  
# Print the result 
print(result) 
```

#### pandas.Series.eq, lt, gt, le,ge, ne

Pandas series.eq() is used to compare every element of Caller series with passed series. It returns True for every element which is Equal to the element in passed series.

```python
a = pd.Series([1, 1, 1, np.nan], index=['a', 'b', 'c', 'd'])
b = pd.Series([1, np.nan, 1, np.nan], index=['a', 'b', 'd', 'e'])
print(a)
print(b)
a.eq(b, fill_value=0)
```

```python
# creating series 1  
series1 = pd.Series([70, 5, 0, 225, 1, 16, np.nan, 10, np.nan])  
     
# creating series 2  
series2 = pd.Series([70, np.nan, 2, 23, 1, 95, 53, 10, 5])  
   
# NaN replacement 
replace_nan = 5
   
# calling and returning to result variable 
result = series1.eq(series2, fill_value = replace_nan) 
   
# display  
result   
```

#### pandas.Series.le

#### idxmax(), idxmin(), sort_values(), sort_index(), head(), tail()

Method to extract the index positions of the lowest values in a Series
