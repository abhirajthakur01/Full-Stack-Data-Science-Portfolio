# Merge multiple excel files into a single file 

```python
import numpy as np
import pandas as pd
```

## Method 1: Using dataframe.append()

```python
# importing the required modules
import glob
```

```python
# Specifying the path to xlsx 
path = "C:\\Users\lenovo\CH - 1\CH5 - Concatenation, Merge\Excel_File1" 
```

```python
# excel files in the path

file_list = glob.glob(path+'/*.xlsx')
```

```python
file_list
```

```python
# list of excel files we want to merge.
# pd.read_excel(file_path) reads the excel
# data into pandas dataframe.

excl_list = []

for file in file_list:
    excl_list.append(pd.read_excel(file))
```

```python
# create a new dataframe to store the 
# merged excel file.
excl_merged = pd.DataFrame()

for excl_file in excl_list:
    excl_merged = excl_merged.append(excl_file,ignore_index=True)
```

```python
excl_merged
```

```python
# exports the dataframe into excel file with
# specified name.
excl_merged.to_excel('total_food_sales.xlsx', index=False)
```

## Method 2: Using pandas.concat()

#### Example:

In the last example, we worked on only two Excel files with a few rows. Let’s try merging more files each containing approximately 5000 rows and 7 columns. We have 5 files BankE, BankD, BankC, BankB, BankA having historical stock data for respective bank. Let’s merge them into a single ‘Bank_Stocks.xlsx’ file. Here we are using the pandas.concat() method.

```python
path1 = "C:\\Users\lenovo\CH - 1\CH5 - Concatenation, Merge\Book"
```

```python
F_List = glob.glob(path1+'/*.xlsx')
```

```python
F_List
```

```python
List1 = []
for file in F_List:
    List1.append(pd.read_excel(file))
```

```python
# concatenate all DataFrames in the list
# into a single DataFrame, returns new 
# DataFrame.
Merged_File = pd.concat(List1, ignore_index=True)
```

```python
Merged_File
```

```python
# exports the dataframe into excel file
# with specified name.
Merged_File.to_excel('Bank_Stocks.xlsx', index=False)
```

### Differences between Concat and Append

Append function will add rows of second data frame to first dataframe iteratively one by one. Concat function will do a single operation to finish the job, which makes it faster than append().

As append will add rows one by one, if the dataframe is significantly very small, then append operation is fine as only a few appends will be done for the number of rows in second dataframe
.
Append function will create a new resultant dataframe instead of modifying the existing one. Due to this buffering and creating process, Append operation’s performance is less than concat() function. However Append() is fine if the number of append operation is a very few. If there are a multiple append operations needed, it is better to use concat().
