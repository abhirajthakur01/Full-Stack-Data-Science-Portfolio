```python
#### Problem:
#You are provided with sales data of various products on Amazon. The company wants to:

#1. Calculate the total sales.
#2. Identify the product with the highest sales.
#3. Find the average sales across products.
#4. Sort the products by their sales amounts.
#5. Apply a sales commission to products and calculate the net sales after commission.
#6. Products with Sales Greater Than $10,000
#7. Products with Sales Less Than or Equal to $8,000
#8. Top 3 Best-Selling Products
#9. Products with Sales Between $6,000 and $15,000
#10. Create a series with missing value of E-Commerce Salse Data
#11. Find out total missing values
#12. Fill missing values with a specific value
#13. Filling missing values with the mean of the series
#14. Fill missing values using forward fill
#15. Fill missing values using backward fill
```

```python
import numpy as np
import pandas as pd
```

```python
# Sales data for Amazon products (fictitious)
amazon_sales_data = {
    'Laptop': 15000,
    'Smartphone': 22000,
    'Headphones': 5000,
    'Camera': 12000,
    'Smartwatch': 8000,
    'Tablet': 14000,
    'Speaker': 6000,
    'Gaming Console': 18000
}    
```

```python
ser = pd.Series(amazon_sales_data)
ser
```

```python
#1 calculate the total sales
Total_Sales = ser.sum()
print("Total Sale Amount:- ",Total_Sales)
```

```python
#2 identify the product with highest sale
Highest_Sale_Prod = ser.idxmax()
Highest_Sale = ser.max()
print(f'Highest Sale Product is : {Highest_Sale_Prod} and Sale is : ${Highest_Sale}')
```

```python
#3 find the average sale across products
Average_Sale = ser.mean()
print(f'Average Sale is : ${Average_Sale}')
```

```python
#4 sort the products by their sales amount
Sorted_Sales = ser.sort_values(ascending=False)
print(f'Products sorted by sales amounts:\n\n{Sorted_Sales}')
```

```python
#5 Apply a sales commission to products and calculate the net sales after commission
com_rate = 0.10
net_sale = ser.apply(lambda x: x*(1-com_rate))
```

```python
print(f'\nNet sales after 10% commission:\n\n{net_sale}')
```

```python
#6. Products with Sales Greater Than $10,000
high_sale_products = ser[ser>10000]
print(f'\nHigh Sale Products are :-\n\n{high_sale_products}')
```

```python
#7 product with less than or equal to 8000
low_sale_products = ser[ser<=8000]
print(f'Low Sale Products are:-\n{low_sale_products}')
```

```python
#8 top 3 best selling products
Top3_BestSale  = ser.sort_values(ascending=False).head(3)
print(f'\nTop Three Best Selling Products Are :- \n\n{Top3_BestSale}')
```

```python
#9 product with sale between 6000 and 15000
Mid_Range = ser[(ser>=6000) & (ser<=15000)]
print(f'\nProducts with sales between $6000 and %15000: \n\n{Mid_Range}')
```

```python
#10 creating a series with missing value
# sales data with missing values (represented by None or NaN)
data = {
    'Laptop': 15000,
    'Smartphone': None,     # Missing sales for Smartphone
    'Headphones': 5000,
    'Camera': 12000,
    'Smartwatch': None,     # Missing sales for Smartwatch
    'Tablet': 14000,
    'Speaker': 6000,
    'Gaming Console': 18000
}

# Create a pandas Series
data = pd.Series(data)

print(data)
```

```python
#11 find out total missing values
total_missing = data.isna().sum()
print(f'Total Missing Values :- {total_missing}')
```

```python
#12 fill missing value with specific value
data1 = data.fillna(0)
print(f'\nFilled missing values with 0:\n\n{data1}')
```

```python
#13 filling the missing value with the mean of the series
data2 = data.fillna(data.mean())
print(f'\nFilled missing values with mean of series:\n\n{data2}')
```

```python
#14 fill missing value using forward fill
data3 = data.ffill()
print(f'\nFilled missing values with mean of series:\n\n{data3}')
```
