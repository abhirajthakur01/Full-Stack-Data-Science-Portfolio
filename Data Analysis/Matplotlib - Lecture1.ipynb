```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### plot temperature values for the given dates

```python
#list storing date in string format
date=["25/12","26/12","27/12"]
#list storing temperature values
temp=[8.5,10.5,6.8]
#create a figure plotting temp versus date
plt.plot(date, temp)
#show the figure
plt.show()
```

```python
date=["25/12","26/12","27/12"]
temp=[8.5,10.5,6.8]
plt.plot(date, temp)
plt.xlabel("Date") #add the Label on x-axis
plt.ylabel("Temperature") #add the Label on y-axis
plt.title("Date wise Temperature") #add the title to the chart
plt.grid(True) #add gridlines to the background
plt.yticks(temp) 
plt.savefig('Temp.png')
plt.show()
```

#### line, line width, marker and marker width

```python
#list storing date in string format
date=["25/12","26/12","27/12"]
#list storing temperature values
temp=[8.5,10.5,6.8]
#create a figure plotting temp versus date
plt.plot(date, temp, color='green',linestyle='--',marker='o', markerfacecolor='blue',markersize = 12, linewidth = 5)
#show the figure
plt.show()
```

```python
#list storing date in string format
date=["25/12","26/12","27/12"]
#list storing temperature values
temp=[8.5,10.5,6.8]
#create a figure plotting temp versus date
plt.plot(date, temp, color='green',linestyle='--',linewidth = 5,marker="D",
         markersize=10, markerfacecolor="midnightblue",
         markeredgewidth=2, markevery=1)
#show the figure
plt.show()
```

```python
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'bs')
plt.plot(t, t**3, 'g^',markevery=5)
plt.show()
```
