# 1. Create a pandas column using for loop

Let’s see how to create a column in pandas dataframe using for loop. Such operation is needed sometimes when we need to process the data of dataframe created earlier for that purpose, we need this type of computation so we can process the existing data and make a separate column to store the data.

It can be easily done by for-loop. The data of column can be taken from the existing Dataframe or any of the array.

```python
import numpy as np
import pandas as pd
```

```python
student = pd.read_csv("students_marks.csv")
```

```python
student.head()
```

```python
student.info()
```

```python
student.columns
```

```python
student[['StudentID','Math']]
```

```python
# create new column Total Marks and Percentage
student['Total_Marks'] = student.Math +student.Science +student.English +student.History + student.Art
student['Percentage'] = round((student.Total_Marks/500)*100,2)
```

```python
student
```

```python
Feedback = []
for i in student.Percentage:
    if i>=75:
        Feedback.append('Hons')
    elif i >=60:
        Feedback.append('First Division')
    elif i>=40:
        Feedback.append('Second Division')
    elif i>=33:
        Feedback.append('Third Division')
    else:
        Feedback.append('Fail')

student['Result'] = Feedback
student
```

```python
student.Result.value_counts().sort_values(ascending=False)
```

```python
student.columns[3:5]
```

## 3. How to rename columns in Pandas DataFrame

Given a Pandas DataFrame, let’s see how to rename columns in Pandas with examples. Here, we will discuss 5 different ways to rename column names in pandas DataFrame.

#### Method 1: Using rename() function

One way of renaming the columns in a Pandas Dataframe is by using the rename() function. This method is quite useful when we need to rename some selected columns because we need to specify information only for the columns which are to be renamed. 

##### Example 1: Rename a single column. 

```python
emp = pd.read_csv("employees_with_errors.csv")
```

```python
emp.head()
```

```python
emp.columns.tolist()
```

```python
emp.rename(columns={'Employe ID' :'Employee_ID'})
```

```python
emp.columns.tolist()
```

```python
emp.rename(columns={'Employe ID' :'Employee_ID'}, inplace=True)
```

```python
emp.columns.tolist()
```

##### Example 2: Rename multiple columns. 

```python
emp.rename(columns={'Nmae':'Name', 'Aeg':'Age'}, inplace=True)
```

```python
emp.columns.tolist()
```

## 6. Change Data Type for one or more columns in Pandas Dataframe

To Change Column Type in Pandas DataFrames, There are different ways of changing DataType for one or more columns in Pandas Dataframe.

```python
#### Change column type into string object using DataFrame.astype() 
```

```python
# sample dataframe
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': ['a', 'b', 'c', 'd', 'e'],
    'C': [1.1, '1.0', '1.3', 2, 5]})
print(df.dtypes)
# converting all columns to string type
df = df.astype(str)
print(df.dtypes)
```

#### Change column type in pandas using DataFrame.apply() 

We can pass pandas.to_numeric, pandas.to_datetime, and pandas.to_timedelta as arguments to apply the apply() function to change the data type of one or more columns to numeric, DateTime, and time delta respectively. 

```python
# sample dataframe
df = pd.DataFrame({
    'A': [1, 2, 3, '4', '5'],
    'B': ['a', 'b', 'c', 'd', 'e'],
    'C': [1.1, '2.1', 3.0, '4.1', '5.1']})
 
# using apply method
df[['A', 'C']] = df[['A', 'C']].apply(pd.to_numeric)
print(df.dtypes)
```
