### Merge Two Files  - Result consist all columns

```python
Result1 = pd.merge(f1,f2,on=['REGISTRATION NO'])
```

```python
Result1
```

```python
# full merge
Result2 = pd.merge(f1,f2,on=['REGISTRATION NO'], how = 'outer')
Result2
```

```python
# Left merge
Result3 = pd.merge(f1,f2,on=['REGISTRATION NO'], how = 'left')
Result3
```

```python
# Right merge
Result4 = pd.merge(f1,f2,on=['REGISTRATION NO'], how = 'right')
Result4
```
