### Creating an array

```python
# Use array function to create an array
data1 = [6,7.5,8,0,1]
arr1 = np.array(data1)
arr1
```

```python
#Nested sequence will be converted into a multidimentional array
data2 = [[1,2,3,4],[5,6,7,8]]
arr2 = np.array(data2)
arr2
```

```python
print("arr1 shape dimention and size")
print(arr1.shape)
print(arr1.ndim)
print(arr1.size)
print(arr1.dtype)
print()
print("arr2 shape dimention and size")
print(arr2.shape)
print(arr2.ndim)
print(arr2.size)
print(arr2.dtype)
```

```python
"""
The numpy.full_like() function is used to create an array 
with the same shape as an existing array and fill it with a 
specific value.
"""
```

```python
a = np.arange(5)
print(np.full_like(a,3))
print(np.full_like(a,1.2))
b = a.astype(float)
print(np.full_like(b,1.2))
print(np.full_like(a,np.nan, dtype=np.double))
```
