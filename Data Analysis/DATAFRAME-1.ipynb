```python
student['Total_Marks'] = student.Math + student.Science  +student.English+student.History +student.Art
```

```python
student['Percentage'] = round((student.Total_Marks/500)*100,2)
```

```python
student
```

```python
student['Percentage'] = round((student.Total_Marks/500)*100,1)
```

```python
student
```

```python
Feedback = []
for i in student.Percentage:
    if i>=75:
        Feedback.append('Hons')
    elif i >= 60:
        Feedback.append('First Division')
    elif i>=40:
        Feedback.append('Second Division')
    elif i>=33:
        Feedback.append('Third Division')
    else:
        Feedback.append('Fail')

student['Result'] = Feedback
student    
```

```python
#Change Data Type for one or more columns in Pandas Dataframe
#To Change Column Type in Pandas DataFrames, There are different ways of changing DataType for one or more columns in Pandas Da
# sample dataframe
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': ['a', 'b', 'c', 'd', 'e'],
    'C': [1.1, '1.0', '1.3', 2, 5]})
print(df.dtypes)
# converting all columns to string type
df = df.astype(str)
print(df.dtypes)
```

```python
# sample dataframe
df = pd.DataFrame({
    'A': [1, 2, 3, '4', '5'],
    'B': ['a', 'b', 'c', 'd', 'e'],
    'C': [1.1, '2.1', 3.0, '4.1', '5.1']})
 
# using apply method
df[['A', 'C']] = df[['A', 'C']].apply(pd.to_numeric)
print(df.dtypes)
```
