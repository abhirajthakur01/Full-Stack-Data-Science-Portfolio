### Method 2 : Query Function

#### This method is elegant and more readable and you don't need to mention dataframe name everytime when you specify columns (variables).

```python
newdf = df.query('origin=="JFK" | carrier == "B6"')
newdf.shape
```

### Method 3 : loc function

```python
newdf = df.loc[(df.origin=='JFK')&(df.carrier=='B6')]
newdf.shape
```

```python
''' loc is an abbreviation of location term. 
All these 3 methods return same output. 
It's just a different ways of doing filtering rows.'''
```

## Filter Pandas Dataframe by Row and Column Position

#### Suppose you want to select specific rows by their position (let's say from second through fifth row). We can use df.iloc[ ] function for the same.

```python
print(df.iloc[:5,]) #First 5 rows
print(df.iloc[1:5,]) #Second to Fifth row
print(df.iloc[5,0]) #Sixth row and 1st column
print(df.iloc[1:5,0]) #Second to Fifth row, first column
print(df.iloc[1:5,:5]) #Second to Fifth row, first 5 columns
print(df.iloc[2:7,1:3]) #Third to Seventh row, 2nd and 3rd column
```

## Selecting multiple values of a column

```python
df.origin.unique()
```

```python
# long way
newdf = df[(df.origin=='JFK')|(df.origin=='LGA')]
print(newdf.shape)
print(newdf.head(2))
```

## How to negate the whole condition

```python
newdf = df[~((df.origin=='JFK')&(df.carrier=='B6'))]
print(newdf.shape)
```

## How to Filter Pandas DataFrames by Column of Strings

## Filtering by single string

#### To filter a DataFrame by a single string value in a column, we can use the str.contains() method. The str.contains() method returns a Boolean mask that can be used to select the rows that contain the specified string value in the column.

```python
import numpy as np
import pandas as pd
```

```python
# create a dataframe
data = {"Name": ["John", "Jane", "Mary", "Adam"],
        "City": ["New York", "Los Angeles", "Chicago", "Houston"]}
df = pd.DataFrame(data)
print(df)
print('\n')

# Filter the DataFrame by a string value in the 'City' column

filtered_df = df[df['City'].str.contains("Los Angeles")]
print(filtered_df)
```

```python
# create a dataframe
data = {"Name": ["John", "Jane", "Mary", "Adam"],
        "City": ["New York", "Los Angeles", "Chicago", "Houston"]}
df = pd.DataFrame(data)
print(df)
print('\n')

# Filter the DataFrame by a string value in the 'City' column

filtered_df = df[df['City']=="Los Angeles"]
print(filtered_df)
```

## Filtering by Multiple Strings

```python
# create a dataframe
data = {"Name": ["John", "Jane", "Mary", "Adam"],
        "City": ["New York", "Los Angeles", "Chicago", "Houston"]}
df = pd.DataFrame(data)
print(df)
print('\n')

# Filter the DataFrame by a string value in the 'City' column

filtered_df = df[df['City'].str.contains("Los Angeles|Chicago")]
print(filtered_df)
```

## Filtering by a List of Strings

```python
# create a dataframe
data = {"Name": ["John", "Jane", "Mary", "Adam"],
        "City": ["New York", "Los Angeles", "Chicago", "Houston"]}
df = pd.DataFrame(data)
print(df)
print('\n')

# Filter the DataFrame by a string value in the 'City' column

filtered_df = df[df['City'].isin(["Los Angeles","Chicago"])]
print(filtered_df)
```
