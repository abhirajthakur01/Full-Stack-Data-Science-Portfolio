```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


```

```python
fig, ax = plt.subplots()
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]

ax.bar(fruits,counts)
ax.set_xlabel('Fruits')
ax.set_ylabel('F   ruits_Count')
ax.set_title('Fruit Supply by Kind and Color')
```

```python
# Method1
fig, ax = plt.subplots()
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels=['red', 'blue', '_red', 'orange']
bar_colors = ['red', 'blue', 'red', 'orange']

ax.bar(fruits,counts,label=bar_labels, color= bar_colors)
ax.set_xlabel('Fruits')
ax.set_ylabel('Fruits_Count')
ax.set_title('Fruit Supply by Kind and Color')
ax.legend()
```

```python
# function to add value labels
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i],ha='center', fontsize=20)
 

# creating data on which bar chart will be plot
x = ["Engineering", "Hotel Management", "MBA",
         "Mass Communication", "BBA", "BSc", "MSc"]
y = [9330, 4050, 3030, 5500,
         8040, 4560, 6650]
     
   
plt.figure(figsize=(14,8))
# making the bar chart on the data
plt.bar(x, y)
     
# calling the function to add value labels
addlabels(x, y)
 

plt.title("College Admission",fontsize=24)
plt.xlabel("Courses",fontsize=22)
plt.ylabel("Number of Admissions",fontsize=22)
plt.xticks(rotation = 45, fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```
