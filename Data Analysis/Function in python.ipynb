```python
#create a function that return both the area and circumference of given circle given radius
import math
def circle_stats(radius):
    area = math.pi*radius**2
    circumference = 2*math.pi*radius
    return area, circumference

a,c = circle_stats(3)
print("area:",a,"circumference:",c)
```

```python
#write a function that greets a user if no name is provided,it should greet with a default name
def greet(name = "User"):
    return "Hello," + name + "!"

print(greet("chai"))
print(greet())
```

```python
#create a function  that takes variable numbers of arguments and return their sum
def sum_all(*args):
    return sum(args)

print(sum_all(1,8,9,4))
print(sum_all(8,7,6,5,4,3,2,1))
```

```python
#create a fun that accepts any num of keyword arguments and print them in format
def print_kwargs(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}: {value}")

print_kwargs(name="shaktiman",power="lazer")
print_kwargs(name="shaktiman")
print_kwargs(name="shaktiman",power="lazer",enemy="dr,jack")
```

```python
#write a generator function that yields even numbers up to a specified limit
def even_generator(limit):
    for i in range(2,limit+1,2):
        yield i

for num in even_generator(10):
    print(num)
```

```python
#create a recursive funstion to calculate the factorialof a num
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
```
