# Pivot

```python
import numpy as np
import pandas as pd
```

```python
data = {
            'Date': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02'],
            'City': ['New York', 'Los Angeles', 'New York', 'Los Angeles'],
            'Temperature': [30, 25, 32, 28]
        }
df=pd.DataFrame(data)
print(df)
print()

pivot_df=df.pivot(index='Date',columns='City',values='Temperature')
print(pivot_df)
```
