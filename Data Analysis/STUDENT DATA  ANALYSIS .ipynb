```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Create the dataset
data = {
    'Student ID': ['S001', 'S002', 'S003', 'S004', 'S005', 'S006', 'S007', 'S008', 'S009', 'S010'],
    'Name': ['Alice', 'Brian', 'Catherine', 'Daniel', 'Emily', 'Frank', 'Grace', 'Henry', 'Isabella', 'Jack'],
    'Age': [16, 17, 16, 15, 17, 16, 15, 17, 16, 15],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'Math Score': [88, 76, 94, 67, 85, 58, 91, 73, 89, 62],
    'English Score': [92, 81, 89, 72, 87, 65, 95, 78, 90, 70],
    'Science Score': [85, 79, 91, 70, 90, 60, 94, 75, 92, 65],
    'Attendance (%)': [95, 88, 98, 80, 93, 76, 99, 85, 97, 82]
}

# Convert to DataFrame
df = pd.DataFrame(data)
```

```python
df
```

```python
#OPERATIONS
print("Mean:\n", df.mean(numeric_only=True))
print("\nMedian:\n", df.median(numeric_only=True))
print("\nMode:\n", df.mode().iloc[0])
print("\nVariance:\n", df.var(numeric_only=True))
print("\nStandard Deviation:\n", df.std(numeric_only=True))
print("\nRange:\n", df.max(numeric_only=True) - df.min(numeric_only=True))
```

```python
avg_scores = df[['Math Score', 'English Score', 'Science Score']].mean()

# Plotting
plt.figure(figsize=(8, 6))
avg_scores.plot(kind='bar', color=['steelblue', 'orange', 'seagreen'])
plt.title('Average Scores by Subject')
plt.ylabel('Average Score')
plt.xlabel('Subject')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Drop non-numeric columns for correlation
numeric_df = df.drop(columns=['Student ID', 'Name', 'Gender'])

# Compute correlation matrix
corr = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f")
plt.title('Heatmap of Student Data Correlation')
plt.tight_layout()
plt.show()
```

```python
# Melt the DataFrame to long format for seaborn
melted_df = pd.melt(df, id_vars=['Name'], 
                    value_vars=['Math Score', 'English Score', 'Science Score'],
                    var_name='Subject', value_name='Score')

# Create the box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=melted_df, x='Subject', y='Score', palette='pastel')
plt.title('Box Plot of Student Scores by Subject')
plt.ylabel('Score')
plt.xlabel('Subject')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
