### Pivot_table and Groupby

#### 1. To calculate maximum and min salary in each department.

```python
df.pivot_table(index='Department',values = 'Salary', aggfunc=[np.max,np.min])
```

```python
#  Or
df.groupby('Department')['Salary'].agg([np.max,np.min])
```

#### 2. To calculate maximum and min salary in each department by position.¶
