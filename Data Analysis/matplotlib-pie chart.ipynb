```python
# Pie chart
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [15, 30, 45, 10]
# only "explode" the 2nd slice (i.e. 'Hogs')
explode = (0, 0.1, 0, 0)  
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.show()
```

```python
# Pie chart
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [15, 30, 45, 10]
# only "explode" the 2nd slice (i.e. 'Hogs')
explode = (0, 0.1, 0, 0)
#add colors
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=90)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Pie chart
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [15, 30, 45, 10]
#colors
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
 
fig1, ax1 = plt.subplots()
patches, texts, autotexts = ax1.pie(sizes, colors = colors, labels=labels, autopct='%1.1f%%', startangle=90)
for text in texts:
    text.set_color('yellow')
for autotext in autotexts:
    autotext.set_color('black')
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.show()
```

```python
df = pd.read_excel("Temp.xlsx")
```

```python
# Plotting for Week1
fig,ax = plt.subplots(figsize=(10,5))
ax.plot(df['Days'].values, df['Week 1'].values)
ax.set_title('A Single Plot')
ax.set_xlabel('Days')
ax.set_ylabel('Temperature')
plt.xticks(rotation=45)
plt.legend('Week1')
```

```python
# Creating three subplots
# Stacking subplots
# Stacking vertically in one direction

fig, axes = plt.subplots(3, figsize=(10,5))
fig.suptitle('Vertically Stacked Subplots')
axes[0].plot(df['Days'].values,df['Week 1'].values,label='sub1')
axes[0].legend()
axes[1].plot(df['Days'].values,df['Week 2'].values,'r', label='sub2')
axes[1].legend()
axes[2].plot(df['Days'].values,df['Week 3'].values, 'y', label='sub3')
axes[2].legend()

plt.tight_layout()
```

```python
# Stacking horizontally in one direction

fig, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(20,3))
l1 = ax1.plot(df['Days'].values,df['Week 1'].values, 'r')
l2 = ax2.plot(df['Days'].values,df['Week 2'].values, 'y')
l3 = ax3.plot(df['Days'].values,df['Week 3'].values, 'b')
ax1.set_xticklabels(df['Days'].values,rotation=45)
ax3.set_xticklabels(df['Days'].values,rotation=90)
lab = ['sub1','sub2','sub3']

fig.legend([l1,l2,l3], labels=lab, loc='upper left')
```

```python
# Stacking horizontally in one direction

fig, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(20,3))
ax1.plot(df['Days'].values,df['Week 1'].values, 'r')
ax2.plot(df['Days'].values,df['Week 2'].values, 'y')
ax3.plot(df['Days'].values,df['Week 3'].values, 'b')
ax1.set_xticklabels(df['Days'].values,rotation=45)
ax3.set_xticklabels(df['Days'].values,rotation=90)
lab = ['sub1','sub2','sub3']

fig.legend([], labels=lab, loc='upper left')
```

```python
fig,ax = plt.subplots(2,2,figsize=(15,4))
fig.suptitle('SUbplots in Two Directions')
ax[0,0].plot(df['Days'].values,df['Week 1'].values)
ax[0,1].plot(df['Days'].values,df['Week 2'].values,'r')
ax[1,0].plot(df['Days'].values,df['Week 3'].values,'g')
ax[1,1].plot(df['Days'].values,df['Week 3'].values,'y')
```

```python
fig, (ax1,ax2) = plt.subplots(2, figsize=(12,3),
                             gridspec_kw={'height_ratios':[2,1]})
ax1.plot(df['Days'].values,df['Week 1'].values)
ax1.set_title('Week1')
ax2.plot(df['Days'].values,df['Week 2'].values,'r')
ax2.set_title('Week 2')

plt.tight_layout()
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(12,3),
                             gridspec_kw={'width_ratios':[2,1]})
ax1.plot(df['Days'].values,df['Week 1'].values)
ax1.set_title('Week1')
ax2.plot(df['Days'].values,df['Week 2'].values,'r')
ax2.set_title('Week 2')

plt.tight_layout()
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(2, figsize=(12,3),
                        gridspec_kw={'height_ratios':[2,1],'hspace':1.4})
ax1.plot(df['Days'].values,df['Week 1'].values)
ax1.set_title('Week1')
ax2.plot(df['Days'].values,df['Week 2'].values,'r')
ax2.set_title('Week 2')

plt.tight_layout()
plt.show()
```
