# Introduction to elementTree

```python
import numpy as np
import pandas as pd
import xml.etree.ElementTree as ET
```

# Parsing XML data 

```python
tree=ET.parse('Movie.xml')
```

```python
tree
```

```python
root=tree.getroot()
```

```python
root
```

```python
root.tag
```

```python
root.attrib
```

# For loop

```python
for child in root:
    print(child.tag,child.attrib)
```

```python
[elem.tag for elem in root.iter()]
```

```python
root.iter()
```

```python
for elem in root.iter():
    print(elem)
```

```python
[movie.attrib for movie in root.iter('movie')]
```

```python
for movie in root.iter('movie'):
    print(movie)
```

```python
for decade in root.iter('decade'):
    print(decade)
```

```python
for format in root.iter('format'):
    print(format)
```
