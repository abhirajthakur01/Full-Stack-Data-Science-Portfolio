### Array Concatenation and Splitting

##### All of the preceding routines worked on single arrays. It’s also possible to combine multiple arrays into one, and to conversely split a single array into multiple arrays. We’ll take a look at those operations here.

##### Concatenation, or joining of two arrays in NumPy, is primarily accomplished through the routines np.concatenate, np.vstack, and np.hstack.

```python
x = np.array([1,2,3])
y = np.array([3,2,1])
np.concatenate([x,y]) # or np.concatenate((x,y))
```

```python
# concatenate more than two arrays at once
z = [99,99,99]
print(np.concatenate([x,y,z]))
```

```python
# two dimensional array
grid = np.array([[1,2,3],[4,5,6]])
grid
```

```python
# concate along the first axis
np.concatenate([grid,grid])
```

```python
# concte along the second axis
np.concatenate([grid,grid],axis=1)
```

#### For working with arrays of mixed dimensions, it can be clearer to use the np.vstack (vertical stack) and np.hstack (horizontal stack) functions:

```python
x = np.array([1,2,3])
grid = np.array([[4,5,6],[7,8,9]])
```

```python
np.concatenate([x,grid])
```

```python
np.vstack([x,grid])
```

```python
np.vstack([grid,x])
```

```python
np.hstack([x,grid])
```

```python
y = np.array([[99],
              [99]])
```

```python
np.hstack([y,grid])
```

```python
np.hstack([grid,y])
```

#### Splitting of arrays

##### The opposite of concatenation is splitting, which is implemented by the functions np.split, np.hsplit, and np.vsplit. For each of these, we can pass a list of indices giving the split points: 

```python
x = [1,2,3,99,99,3,2,1]
x1,x2,x3 = np.split(x,[3,5]) #3 and 5 are indices
print(x1,x2,x3)
```

```python
# Notice that N split points lead to N + 1 subarrays. The related functions np.hsplit
# and np.vsplit are similar:
```

```python
grid = np.arange(16).reshape(4,4)
grid
```

```python
upper,lower = np.vsplit(grid,[1]) # 1 is indice
print(upper,'\n','\n',lower)
```

```python
upper,lower = np.vsplit(grid,[3])
print(upper,'\n','\n',lower)
```

```python
left, right = np.hsplit(grid,[2])
print(left,'\n ',' \n',right)
```
