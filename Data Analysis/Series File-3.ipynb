```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Math': [85, 92, 78, 90, 88],
    'Science': [91, 85, 89, 92, 94],
    'English': [78, 82, 84, 88, 90]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
#extract series from data frame
math_series = df['Math']
print(math_series)
```

```python
#Find out mean, max, min, std and median of math_series
print('Mean:- ', math_series.mean())
print('Mean:- ', math_series.max())
print('Mean:- ', math_series.min())
print('Mean:- ', math_series.median())
print('Mean:- ', math_series.mode())
```

```python
#Students who scored more than 85 in Math
print('No. of students:-', math_series[math_series>85])
```
