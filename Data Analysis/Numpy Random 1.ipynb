### Shuffle The Elements Of The List

#### shuffle() changing the arrangement of the elements in-place such that the original array is changed.

#### permutation() The permutation() method returns a re-arranged array such that the original array is unchanged.

#### shuffle()

```python
mylist = np.array(['USA', 'India', 'China', 'Dubai'])
np.random.shuffle(mylist)
print(mylist)
```

#### permutaion()

```python
mylist = np.array(['USA', 'India', 'China', 'Dubai'])
print(np.random.permutation(mylist))
print(mylist)
```

### Example1:- Write a NumPy program to shuffle numbers between 0 and 10 (inclusive).

```python
# Generating an array 'x' with numbers from 0 to 9 using np.arange()
x = np.arange(10)
print(x)
# Shuffling the elements of the array 'x' in-place using np.random.shuffle()
np.random.shuffle(x)

# Printing the shuffled array 'x'
print(x)

# Generating a permutation of numbers from 0 to 9 using np.random.permutation()
# and printing the result
print("Same result using permutation():")
print(x)
print(np.random.permutation(x))
print(x)
```

### Example3- Write a NumPy program to create a 5x5 array with random values and find the minimum and maximum values.

```python
# Generating a 5x5 NumPy array 'x' filled with random floats between 0 and 1 using np.random.random()

x = np.random.random((5, 5))

# Printing the original array 'x'
print("Original Array:")
print(x)

# Calculating the minimum and maximum values in array 'x' using x.min() and x.max() functions
xmin, xmax = x.min(), x.max()

# Printing the minimum and maximum values of the array 'x'
print("Minimum and Maximum Values:")
print(xmin, xmax) 
```

### Example4 -Write a NumPy program to create a random 10x4 array and extract the first five rows of the array and store them into a variable.

### Example6 - Write a NumPy program to check two random arrays are equal or not.

```python
# Generating the first array 'x' with 6 random integers (0 or 1) using np.random.randint()
x = np.random.randint(0, 2, 6)

# Printing the first generated array 'x'
print("First array:")
print(x)

# Generating the second array 'y' with 6 random integers (0 or 1) using np.random.randint()
y = np.random.randint(0, 2, 6)

# Printing the second generated array 'y'
print("Second array:")
print(y)

# Checking if the two arrays 'x' and 'y' are equal using np.allclose()
# This function returns True if the two arrays are element-wise equal within a tolerance
array_equal = np.allclose(x, y)

# Printing whether the two arrays are equal or not
print("Test above two arrays are equal or not!")
print(array_equal)
```

### Example7 - Write a NumPy program to create random vector of size 15 and replace the maximum value by -1.

```python
# Generating an array 'x' with 15 random floats between 0 and 1 using np.random.random()
x = np.random.random(15)

# Displaying the original array 'x'
print("Original array:")
print(x)

# Finding the index of the maximum value in the array using x.argmax()
# Replacing the maximum value with -1 using the index obtained from x.argmax()
x[x.argmax()] = -1

# Displaying the modified array where the maximum value has been replaced by -1
print("Maximum value replaced by -1:")
print(x) 
```

```python
# Generating an array 'x' with 15 random floats between 0 and 1 using np.random.random()
x = np.random.randint(15, size=15)

# Displaying the original array 'x'
print("Original array:")
print(x)
print(np.argmax(x))

# Finding the index of the maximum value in the array using x.argmax()
# Replacing the maximum value with -1 using the index obtained from x.argmax()
x[np.argmax(x)] = -1

# Displaying the modified array where the maximum value has been replaced by -1
print("Maximum value replaced by -1:")
print(x) 
```
