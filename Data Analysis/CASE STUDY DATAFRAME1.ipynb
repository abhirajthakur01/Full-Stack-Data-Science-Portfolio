```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel("Case Study 1 - DataFrame.xlsx")
```

```python
df
```

```python
#change the name of all columns
Col = ['Sno','District','A-Wheat','A-Rice','A-Sugar','A-Salt','A-FSalt','A-FRice',
       'RO-Wheat','RO-Rice','RO-Sugar','RO-Salt','RO-FSalt','RO-FRice',
       'D-Wheat','D-Rice','D-Sugar','D-Salt','D-FSalt','D-FRice',
       'R-Wheat','R-Rice','R-Sugar','R-Salt','R-FSalt','R-FRice']
```

```python
df.columns = Col
```

```python
df.head()
```

```python
#set Sno columns as index columns
df.set_index('Sno', inplace=True)
```

```python
df.head()
```

```python
df.columns
```

```python
#3 drop all columns which start from RO
df.drop(['RO-Wheat', 'RO-Rice', 'RO-Sugar', 'RO-Salt', 'RO-FSalt',
       'RO-FRice'],axis=1,inplace=True)
```

```python
df.columns
```

```python
#4 add values of same commodity
df['A-RICE'] = df['A-Rice']+df['A-FRice']
df['D-RICE'] = df['D-Rice']+df['D-FRice']
df['R-RICE'] = df['R-Rice']+df['R-FRice']

df['A-SALT'] = df['A-Salt']+df['A-FSalt']
df['D-SALT'] = df['D-Salt']+df['D-FSalt']
df['R-SALT'] = df['R-Salt']+df['R-FSalt']
```

```python
df.columns
```

```python
#5 drop columns which values added in new columns
df.drop(['A-Rice','A-Salt','A-FSalt','A-FRice','D-Rice', 'D-Salt', 'D-FSalt',
       'D-FRice', 'R-Rice', 'R-Salt', 'R-FSalt','R-FRice'],axis=1,inplace=True)
```

```python
df.columns
```

```python
#6 arrange all the columns in sequence
df = df[['District', 'A-Wheat', 'A-RICE', 'A-SALT', 'A-Sugar', 'D-Wheat','D-RICE', 'D-Sugar', 'D-SALT', 'R-Wheat',
       'R-Sugar',  'R-RICE', 'R-SALT']]
```

```python
df.columns
```

```python
L1 = list(df.columns)
```

```python
L1
```

```python
### Q7 Insert D-%W, D-%R,D-%S,D-%FS column at index 9,10,11,12
for i,j in enumerate(L1):
    print(i, " ",j)
```

```python
df.insert(loc = 9, column='D-%W', value=round(df['D-Wheat']/df['A-Wheat']*100,0))
```

```python
df.head(2)
```

```python
for i,j in enumerate(list(df.columns)):
    print(i, " ",j)
```

```python
df.insert(loc = 10, column='D-%R', value=round(df['D-RICE']/df['A-RICE']*100,0))
df.insert(loc = 11, column='D-%S', value=round(df['D-Sugar']/df['A-Sugar']*100,0))
df.insert(loc = 12, column='D-%FS', value=round(df['D-SALT']/df['A-SALT']*100,0))
```

```python
df.head()
```

```python
for i,j in enumerate(list(df.columns)):
    print(i, " ",j)
```

```python
#Q8 Insert R-%W, R-%R,R-%S,R-%FS column at index 17,18,19,20¶
df.insert(loc = 17, column='R-%W', value=round(df['R-Wheat']/df['A-Wheat']*100,0))
df.insert(loc = 18, column='R-%R', value=round(df['R-RICE']/df['A-RICE']*100,0))
df.insert(loc = 19, column='R-%S', value=round(df['R-Sugar']/df['A-Sugar']*100,0))
df.insert(loc = 20, column='R-%FS', value=round(df['R-SALT']/df['A-SALT']*100,0))
```

```python
df.columns
```

```python
df1 = df[['District', 'A-Wheat', 'A-RICE', 'A-SALT', 'A-Sugar',  
          'D-%W', 'D-%R', 'D-%S', 'D-%FS',
           'R-%W', 'R-%R', 'R-%S','R-%FS']]
```

```python
for i,j in enumerate (list(df1.columns)):
    print(i," ",j)
```

```python
### Q9 Insert Avg-D-%WR and Avg-R-%WR columns at index 7 and 12
df1.insert(loc=7, column="Avg-D-%WR", value=round(((df1['D-%W']+df1['D-%R'])/2),0))
df1.insert(loc=12, column="Avg-R-%WR", value=round(((df1['R-%W']+df1['R-%R'])/2),0))
```

```python
df1.sort_values('Avg-R-%WR', ignore_index=True).head(2)
```

```python
df1['A-Wheat'] = round((df1['A-Wheat'])/1000,0)
df1['A-RICE'] = round((df1['A-RICE'])/1000,0)
df1['A-SALT'] = round((df1['A-SALT'])/1000,0)
df1['A-Sugar'] = round((df1['A-Sugar'])/1000,0)
```
