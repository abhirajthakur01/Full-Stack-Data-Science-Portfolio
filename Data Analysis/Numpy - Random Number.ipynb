# Random Numbers in Numpy

#### Random number - that can not be predicted logically

```python
import numpy as np
import pandas as pd
```

### Generate a random integer

```python
#Generate a random integer from 0 to 50
n = np.random.randint(50)
print(n)
```

```python
#Generate a random integer from 1 to 10
n = np.random.randint(1,10)
print(n)
```

### Generate random array

```python
# Generate a one dimension array containing 10 intergers from 0 to 100
n = np.random.randint(100, size=(10))
n
```

```python
# Generate a two dimension array containing 4 rows and each row contain 6 random intergers from 0 to 100
n = np.random.randint(100, size=(4,5))
n
```

```python
# Generate a one dimesional arrary containing 10 random integers
# rand function generate floating type random number between  0 and 1
n = np.random.rand(10)
n
```

```python
#Generate a two dimenstional array with 5 rows, each row containing 6 floating point random numbers between 0 and 1:
n = np.random.rand(5,6)
n
```

### Generate random number from array

```python
#The choice() method takes an array as a parameter and return random value from array.
n = np.random.choice([3,5,7,9])
n
```

```python
# Generate a 2-D array that consists of the values in the array parameter (3, 5, 7, and 9):
n = np.random.choice([3,5,7,9], size=(3,5))
n
```
