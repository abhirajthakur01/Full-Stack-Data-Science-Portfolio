```python
faqs = """What is the purpose of the Student Focus Initiative?
The initiative is designed to help students build consistent study habits, manage academic pressure, and improve daily learning routines.

Who can participate in this initiative?
Any student from high school or college can take part, regardless of stream or academic performance.

How long does the initiative run?
The initiative runs for five months, with activities planned weekly.

What kind of activities are included?
Students engage in:

Daily study planning exercises

Memory and recall practices

Time-management challenges

Peer discussion circles

Concept-revision techniques

Stress-handling strategies

Self-reflection journals

Monthly progress reviews

Does this initiative focus on exam preparation?
No, the focus is on learning behavior and consistency, not direct exam coaching.

What if a student misses a scheduled activity?
Missed activities can be completed later using the shared resources.

Where can students find the weekly plan?
The weekly roadmap is available on the student dashboard.

How much time does a student need to spend daily?
Most activities require 30вҖ“45 minutes per day.

Which language is used in discussions and materials?
Content is shared in simple English with occasional Hindi phrases.

How are students informed about upcoming tasks?
Students receive notifications through email and in-app alerts.

Is this suitable for students who struggle with concentration?
Yes, the initiative is especially helpful for students facing focus or motivation issues.

Can a student join after the initiative has begun?
Yes, late entries are allowed.

Will previous materials be accessible to late joiners?
Yes, all earlier activities and resources remain accessible.

Are students evaluated or graded?
No formal grading is involved. Progress is self-monitored.

Are real student examples discussed?
Yes, anonymized student experiences are shared during group sessions.

How can students ask questions or seek guidance?
Students can submit queries through a feedback form, and mentors respond individually.
"""
```

```python
# OHE
# Integer Encoding

Corpus

1: data
2: Vocabulary
3: Encoding
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokens=Tokenizer()
```

```python
tokens.fit_on_texts([faqs])
```

```python
tokens.word_index
```

```python
input_sequences=[]
for s in faqs.split("\n"):
    tokenized_sentence=tokens.texts_to_sequences([s])[0]

    for i in range(1,len(tokenized_sentence)):
        input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len=max([len(x) for x in input_sequences])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
padded_input_sequences=pad_sequences(input_sequences,maxlen=max_len,padding="pre")
```

```python
padded_input_sequences
```

```python
x=padded_input_sequences[:,:-1]
y=padded_input_sequences[:,-1]
```

```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=283)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
```

```python
model = Sequential()
model.add(Embedding(283, 100, input_shape=(max_len,)))
model.add(LSTM(150, dropout=0.2))
model.add(Dense(283, activation="softmax"))
```

```python
model.compile(loss="categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
model.summary()
```

```python
model.fit(x,y,epochs=100)
```

```python
import numpy as np
```

```python
text="which language is"

for i in range(10):
    token_text=tokens.texts_to_sequences([text])[0]
    padded_token_text=pad_sequences([token_text],maxlen=56,padding="pre")
    pos=np.argmax(model.predict(padded_token_text))
    for word,index in tokens.word_index.items():
        if index==pos:
            text=text+" "+word
            print(text)
            
```
