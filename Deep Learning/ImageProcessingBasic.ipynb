```python
# What is Stacking in image processing
In image processing, stacking refers to the process of combining multiple images, often of the same scene, to produce a single, enhanced image. 
This technique is commonly used to improve image quality by reducing noise, enhancing details, and increasing dynamic range.

Key Purposes of Stacking in Image Processing
1: Noise Reduction:
When multiple images of the same scene are taken, they will contain slight variations due to noise. 
By averaging or combining these images, random noise can be reduced, resulting in a clearer image.

2: Detail Enhancement:
Stacking can help bring out finer details that may be lost in a single image due to low light or camera limitations. 
This is common in astrophotography, where stacking multiple images of stars can reveal details that aren’t visible in a single shot.

3: Dynamic Range Improvement: 
Stacking images taken at different exposures allows for a wider dynamic range, which means better visibility of both highlights and shadows. 
This is used in High Dynamic Range (HDR) imaging.

Types of Stacking Methods
Averaging: 
Each pixel’s value is averaged across multiple images, reducing random noise.
Median Stacking: 
The median value of each pixel across images is taken. This is effective for removing transient objects like moving objects or noise.
Max or Min Stacking: 
The maximum or minimum values of each pixel are taken, often used to highlight certain features in a sequence of images.
```

```python
# stacking of images 
import cv2
import numpy as np

# Load multiple images of the same scene into a list
images = [cv2.imread(f'image{i}.jpg') for i in range(1, 3)]

# Stack images using averaging
stacked_image = np.mean(images, axis=0).astype(np.uint8)

# Display result
cv2.imshow("Stacked Image", stacked_image)
cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Create a blank black image
img = cv2.imread('flower.png')

# Define text and position
text = "Hello, India!"
position = (50, 250)

# Add white text to the image
cv2.putText(img, text, position, cv2.FONT_HERSHEY_SIMPLEX, 1, 
            (255, 255, 255), 2, cv2.LINE_AA)

# Display the image
cv2.imshow('Image with Text', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img: 
The image on which the text will be placed.
text: 
The string of text you want to display.
org: 
The bottom-left corner of the text in (x, y) format.
fontFace: 
The font type. Some common options:
cv2.FONT_HERSHEY_SIMPLEX
cv2.FONT_HERSHEY_PLAIN
cv2.FONT_HERSHEY_DUPLEX
cv2.FONT_HERSHEY_COMPLEX
cv2.FONT_HERSHEY_TRIPLEX
cv2.FONT_HERSHEY_COMPLEX_SMALL
cv2.FONT_HERSHEY_SCRIPT_SIMPLEX
cv2.FONT_HERSHEY_SCRIPT_COMPLEX
cv2.FONT_HERSHEY_COMPLEX | cv2.FONT_ITALIC (to make italic)

fontScale: 
The scale factor that determines the size of the text.
color: 
Text color in BGR format, e.g., (255, 0, 0) for blue.
thickness: 
Thickness of the text strokes.
lineType: 
Type of line used to draw the text. Options include:
cv2.LINE_4 (for a 4-connected line)
cv2.LINE_8 (for an 8-connected line, default)
cv2.LINE_AA (for anti-aliased, smoother text)

bottomLeftOrigin: 
When true, the text origin is at the bottom-left corner. By default, it’s set to False.
```

```python
# Event Handling part-1
img=np.zeros((512,512,3))
def action1(event,x,y,flags,params):
    print("Event Triggered")
def action2(event,x,y,flags,params):
    if event==0:
        print("Mouse Move")
    elif event==1:
        print("Mouse  Clicked")
    elif event==4:
        print("Mouse Click Drop")
def action3(event,x,y,flags,params):
    if event==1:
        cv2.circle(img,center=(x,y),radius=30,color=(0,0,255),thickness=2)
cv2.namedWindow(winname="window")
cv2.setMouseCallback("window",action2)
while True:
    cv2.imshow("window",img)
    if cv2.waitKey(1) & 0xFF==ord("s"):
        break
cv2.destroyAllWindows()
```

```python
# Event Handling Part 2:
img=np.zeros((512,512,3))
draw=False
ix=-1
iy=-1
def action(event,x,y,flags,params):
    global draw,ix,iy
    if event==1:
        draw=True
        ix=x
        iy=y
    elif event==0:
        if draw==True:
            cv2.rectangle(img,pt1=(ix,iy),pt2=(x,y),color=(0,255,255),thickness=-1)
    elif event==4:
        draw=False
        cv2.rectangle(img,pt1=(ix,iy),pt2=(x,y),color=(0,255,255),thickness=-1)   
cv2.namedWindow(winname="window")
cv2.setMouseCallback("window",action)
while True:
    cv2.imshow("window",img)
    if cv2.waitKey(1) & 0xFF==ord("s"):
        break
cv2.destroyAllWindows()


```

```python
# Event Handling part 3
import cv2
import numpy as np
# Callback function for mouse events
def draw_circle(event, x, y, flags, param):
    # Check for left button down event to draw a circle
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img, (x, y), 10, (255, 0, 0), -1)  # Blue circle on left-click
    # Check for right button down event to draw a rectangle
    elif event == cv2.EVENT_RBUTTONDOWN:
        cv2.rectangle(img, (x - 20, y - 20), (x + 20, y + 20), (0, 255, 0), -1)  # Green square on right-click
img = np.zeros((512, 512, 3), dtype=np.uint8)
# Create a window and set the mouse callback function
cv2.namedWindow('Event Handling')
cv2.setMouseCallback('Event Handling', draw_circle)
while True:
    # Display the image
    cv2.imshow('Event Handling', img)
    # Break the loop on pressing the 'ESC' key
    if cv2.waitKey(1) & 0xFF == 27:
        break
# Clean up
cv2.destroyAllWindows()
```

```python
# Cropping Tool 
import cv2
cropping = False
x_start, y_start, x_end, y_end = 0, 0, 0, 0
def crop_event_handler(event, x, y, flags, param):
    global x_start, y_start, x_end, y_end, cropping
    # Start cropping on left mouse button down
    if event == cv2.EVENT_LBUTTONDOWN:
        cropping = True
        x_start, y_start = x, y
    # Update end point as the mouse moves
    elif event == cv2.EVENT_MOUSEMOVE:
        if cropping:
            x_end, y_end = x, y
    # Complete cropping on left mouse button release
    elif event == cv2.EVENT_LBUTTONUP:
        cropping = False
        x_end, y_end = x, y
        # Crop and display the selected ROI
        if x_start != x_end and y_start != y_end:
            roi = image[y_start:y_end, x_start:x_end]
            cv2.imshow("Cropped Image", roi)
image = cv2.imread('flower.png')
clone = image.copy()
# Set up the window and callback function
cv2.namedWindow("Image")
cv2.setMouseCallback("Image", crop_event_handler)

while True:
    # Display the image and draw rectangle while cropping
    temp_image = clone.copy()
    if cropping:
        cv2.rectangle(temp_image, (x_start, y_start), (x_end, y_end), (0, 255, 0), 2)
    
    cv2.imshow("Image", temp_image)

    # Exit on pressing 'q'
    key = cv2.waitKey(1) & 0xFF
    if key == ord("s"):
        break

# Clean up
cv2.destroyAllWindows()
```

```python
# Resize Tool in CV
import cv2
image = cv2.imread('flower.png')  
clone = image.copy()
# Function to resize the image based on scale from trackbar
def resize_image(scale):
    scale = max(scale, 1)  # Prevent scale from going to 0 or negative
    width = int(clone.shape[1] * scale / 100)
    height = int(clone.shape[0] * scale / 100)
    resized = cv2.resize(clone, (width, height), interpolation=cv2.INTER_LINEAR)
    cv2.imshow("Resize Tool", resized)
# Initialize window and trackbar
cv2.namedWindow("Resize Tool")
cv2.createTrackbar("Scale %", "Resize Tool", 100, 200, resize_image)  # Default scale at 100%
# Display the initial image
resize_image(100)  # Show the image at 100% scale
# Keep the window open until 'q' is pressed
while True:
    if cv2.waitKey(1) & 0xFF == ord('s'):
        break
# Clean up
cv2.destroyAllWindows()
```

```python
# switching color and grayscale onclick
import cv2
image = cv2.imread('flower.png')  
grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
click_count = 0
def toggle_grayscale(event, x, y, flags, param):
    global click_count
    if event == cv2.EVENT_LBUTTONDOWN:
        click_count += 1
        # Display grayscale on even clicks, color on odd clicks
        if click_count % 2 == 0:
            cv2.imshow("Toggle Grayscale", grayscale_image)
        else:
            cv2.imshow("Toggle Grayscale", image)
# Create a window and set the mouse callback function
cv2.namedWindow("Toggle Grayscale")
cv2.setMouseCallback("Toggle Grayscale", toggle_grayscale)
# Display the original image initially
cv2.imshow("Toggle Grayscale", image)
# Keep the window open until 'q' is pressed
while True:
    if cv2.waitKey(1) & 0xFF == ord('s'):
        break
# Clean up
cv2.destroyAllWindows()
```

```python
# on your web cam
import cv2
# Open the default webcam (usually index 0)
cap = cv2.VideoCapture(0)
# Check if the webcam is opened successfully
if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()
# Read and display frames in a loop
while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    # If frame is read correctly, ret is True
    if not ret:
        print("Error: Could not read frame.")
        break
    # Display the frame
    frame = cv2.flip(frame, 1)  # to correct horizontal flipping
    cv2.imshow('Webcam', frame) 
    # Exit the loop when 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('s'):
        break
# Release the webcam and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```

```python
# Video Recording using web cam
import cv2
# Open the default webcam
cap = cv2.VideoCapture(0)
# Check if the webcam is opened successfully
if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()
# Define the codec and create VideoWriter object
# 'XVID' codec is used, and the output is saved as 'output.avi'
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi', fourcc, 20.0, (640, 480))
# Capture and save video frames in a loop
while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    # If frame is read correctly, ret is True
    if not ret:
        print("Error: Could not read frame.")
        break
    # Write the frame to the output file
    out.write(frame)
    # Display the frame
    cv2.imshow('Recording', frame)
    if cv2.waitKey(1) & 0xFF == ord('s'):
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```
