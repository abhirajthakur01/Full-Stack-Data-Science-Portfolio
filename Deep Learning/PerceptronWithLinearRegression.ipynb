## Simple Linear Regression Using single Manual Perceptron

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some synthetic data (linear relationship)
np.random.seed(42)  
X = 2 * np.random.rand(100, 1)  # 100 random data points for X between 0 and 2
y = 3 * X + 5 + np.random.randn(100, 1)  # y = 3x + 5 + some noise

# Plot the dataset
plt.scatter(X, y, color='blue')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Synthetic Data for Linear Regression')
plt.show()

# Perceptron as Linear Regression
class LinearPerceptron:
    def __init__(self, learning_rate=0.01, n_iterations=1000):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.weights = None
        self.bias = None
    def fit(self, X, y):
        n_samples, n_features = X.shape
        # Initialize weights and bias
        self.weights = np.zeros((n_features, 1))
        self.bias = 0
        # Gradient Descent
        for _ in range(self.n_iterations):
            y_predicted = np.dot(X, self.weights) + self.bias
            # Compute gradients
            dw = (1 / n_samples) * np.dot(X.T, (y_predicted - y))
            db = (1 / n_samples) * np.sum(y_predicted - y)
            # Update weights and bias
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db
    
    def predict(self, X):
        return np.dot(X, self.weights) + self.bias

# Initialize the linear perceptron model
model = LinearPerceptron(learning_rate=0.01, n_iterations=1000)

# Train the model on the synthetic dataset
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Plot the predictions against the original data
plt.scatter(X, y, color='blue', label='Original data')
plt.plot(X, y_pred, color='red', label='Fitted line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression using Perceptron')
plt.legend()
plt.show()

# Print the learned weights and bias
print(f"Learned weight: {model.weights[0][0]}")
print(f"Learned bias: {model.bias}")
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt

# Generate dummy data for simple linear regression
np.random.seed(0)
X = np.linspace(0, 10, 100).reshape(-1, 1)
y = 2.5 * X.flatten() + 4 + np.random.randn(100)  # y = 2.5x + 4 + noise

# Build a simple ANN with 1 neuron (no hidden layer)
model = Sequential()
model.add(Dense(units=1, input_dim=1, activation='linear'))

# Compile with MSE loss for regression
model.compile(optimizer='adam', loss='mse')

# Train the model
model.fit(X, y, epochs=1000)

# Predict
y_pred = model.predict(X)

# Plot
plt.scatter(X, y, label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.legend()
plt.title("Simple Linear Regression using ANN")
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_absolute_error

# After training
y_pred = model.predict(X)

# Mean Absolute Error
mae = mean_absolute_error(y, y_pred)
print(f"Mean Absolute Error: {mae:.4f}")

# R2 Score
r2 = r2_score(y, y_pred)
print(f"RВІ Score: {r2:.4f}")
```
