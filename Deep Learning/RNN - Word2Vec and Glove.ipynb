```python

import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from gensim.models import KeyedVectors

sentences = [
    "I love this course",
    "This is amazing",
    "I hate this topic",
    "This is boring"
]
labels = np.array([1, 1, 0, 0])

tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
sequences = tokenizer.texts_to_sequences(sentences)
padded = pad_sequences(sequences, padding='post')
vocab_size = len(tokenizer.word_index) + 1
max_len = padded.shape[1]


# WORD2VEC EMBEDDING

# Load pre-trained Word2Vec (Google News, 300-dim)
# word2vec = KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin', binary=True)
# For demo purposes, we simulate small vectors
embedding_dim = 50
embedding_matrix_w2v = np.random.rand(vocab_size, embedding_dim)

# Fill embedding_matrix_w2v with Word2Vec vectors if available
# for word, idx in tokenizer.word_index.items():
#     if word in word2vec:
#         embedding_matrix_w2v[idx] = word2vec[word]

# Build RNN model with Word2Vec embeddings
model_w2v = Sequential()
model_w2v.add(Embedding(input_dim=vocab_size,
                        output_dim=embedding_dim,
                        weights=[embedding_matrix_w2v],
                        input_length=max_len,
                        trainable=False))  # freeze pre-trained embeddings
model_w2v.add(SimpleRNN(10))
model_w2v.add(Dense(1, activation='sigmoid'))

model_w2v.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
print("Training RNN with Word2Vec Embeddings")
model_w2v.fit(padded, labels, epochs=20, verbose=0)
```

```python

# GLOVE EMBEDDING

# Load pre-trained GloVe vectors (e.g., 'glove.6B.50d.txt')
# For demo, simulate small vectors
embedding_matrix_glove = np.random.rand(vocab_size, embedding_dim)

# Build RNN model with GloVe embeddings
model_glove = Sequential()
model_glove.add(Embedding(input_dim=vocab_size,
                          output_dim=embedding_dim,
                          weights=[embedding_matrix_glove],
                          input_length=max_len,
                          trainable=False))
model_glove.add(SimpleRNN(10))
model_glove.add(Dense(1, activation='sigmoid'))

model_glove.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
print("\nTraining RNN with GloVe Embeddings")
model_glove.fit(padded, labels, epochs=20, verbose=0)
```
