```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
dst = pd.read_csv("Marksheet.csv") #data collecting
```

```python
dst
```

```python
x = pd.DataFrame(dst.iloc[:,2:4])  #independent column
y = pd.DataFrame(dst.iloc[:,-1]) #dependent column
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=1/3,random_state= 0) #split the datatset
```

```python
#fitting the linear regression model to the trainging dataset

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#prediction of test data and training data set result
x_test_pred = regressor.predict(x_test)
from sklearn.metrics import r2_score
r2_score(y_test,x_test_pred)
```
