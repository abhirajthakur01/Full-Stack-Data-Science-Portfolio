```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score

# Generate synthetic data
X = 6 * np.random.rand(200, 1) - 3
y = 0.8 * X**2 + 0.9 * X + 2 + np.random.randn(200, 1)

# Visualize the data
plt.plot(X,y,'b.')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Polynomial Regression Example")
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)

# Create model and train
lr = LinearRegression()
lr.fit(X_train, y_train)

# Predict
y_pred = lr.predict(X_test)

# Check accuracy (R² score)
print("R2 Score:", r2_score(y_test, y_pred))

# Plot results
plt.scatter(X, y, color="b")  # Original data
plt.plot(X_train, lr.predict(X_train), color="r")  # Regression line
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
poly = PolynomialFeatures(degree=2, include_bias=True)
X_train_trans = poly.fit_transform(X_train)
X_test_trans = poly.transform(X_test)

# Fit model
lr = LinearRegression()
lr.fit(X_train_trans, y_train)

# Predictions
y_pred = lr.predict(X_test_trans)
print("R² score:", r2_score(y_test, y_pred))

# Visualization
X_new = np.linspace(-3, 3, 200).reshape(200, 1)
X_new_poly = poly.transform(X_new)
y_new = lr.predict(X_new_poly)
plt.plot(X_new, y_new, color="red", linewidth=2, label="Polynomial Regression")
plt.plot(X_train,y_train,'b.',label='training points')
plt.plot(X_test,y_test,'g.',label='training points')
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
