## Bagging in Ensemble Learning

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier, VotingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# 1. Load dataset
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 2. Define different base learners
dt = DecisionTreeClassifier(random_state=42)
knn = KNeighborsClassifier()
lr = LogisticRegression(max_iter=1000)

# 3. Wrap each base learner with Bagging (use 'estimator' instead of 'base_estimator')
bag_dt = BaggingClassifier(estimator=dt, n_estimators=10, random_state=42)
bag_knn = BaggingClassifier(estimator=knn, n_estimators=10, random_state=42)
bag_lr = BaggingClassifier(estimator=lr, n_estimators=10, random_state=42)

# 4. Combine them into one ensemble
ensemble = VotingClassifier(
    estimators=[
        ('bag_dt', bag_dt),
        ('bag_knn', bag_knn),
        ('bag_lr', bag_lr)
    ],
    voting='soft'
)

# 5. Train and evaluate
ensemble.fit(X_train, y_train)
y_pred = ensemble.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```
