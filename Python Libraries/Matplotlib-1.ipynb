```python
x = np.arange(10,20)
y = np.arange(100,90,-1)
z = np.arange(-10,0)
```

```python
plt.plot(x,y)
plt.show()
plt.plot(x,z)
plt.show()
```

```python
plt.subplot(1,2,1)        #subplot parameters (no.of row,no. of columns,position of the belove plot)  it is used to plot multiple graph in same row
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(x,z)
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,z)
plt.subplot(2,2,3)
plt.plot(x,z)
plt.subplot(2,2,4)
plt.plot(z,y)
plt.show()
```

```python
plt.subplot(2,2,1)
plt.title("1st plot")
plt.plot(x,y)
plt.subplot(2,2,2)
plt.title("2nd plot")
plt.plot(x,z)
plt.subplot(2,2,3)
plt.title("3nd plot")
plt.xlabel("age")
plt.ylabel("salary")
plt.plot(x,z)
plt.subplot(2,2,4)
plt.title("4nd plot")
plt.tight_layout() #to prevent overlapping
plt.plot(z,y)
plt.show()
```

```python
plt.subplot(2,2,1)
plt.title("1st plot")
plt.plot(x,y)
plt.subplot(2,2,2)
plt.title("2nd plot")
plt.plot(x,z)
plt.subplot(2,2,3)
plt.title("3nd plot")
plt.xlabel("age")
plt.ylabel("salary")
plt.plot(x,z)
plt.subplot(2,2,4)
plt.title("4nd plot")
plt.tight_layout() #to prevent overlapping
plt.suptitle("Title of all the plot")
plt.plot(z,y)
plt.show()
```

```python
x = [1,21,31,12,51,67,87,100]
y = [-2,22,-45,-12,32,98,-1,2]
colors = x
plt.scatter(x,y, c=colors,cmap = "afmhot")
plt.colorbar()
plt.show()
```

```python
x = [1,21,31,12,51,67,87,100]
y = [-2,22,-45,-12,32,98,-1,2]
colors = x
sizes = [100,21,31,12,51,67,87,100]
plt.scatter(x,y, c=colors,cmap = "coolwarm",s = sizes)
plt.colorbar()
plt.show()
```

```python
x = [1,21,31,12,51,67,87,100]
y = [-2,22,-45,-12,32,98,-1,2]
colors = x
sizes = [1000,21,31,12,51,67,87,100]
plt.scatter(x,y, c=colors,cmap = "nipy_spectral",s = sizes,alpha = 0.5) #alpha for transperancy value
plt.colorbar()
plt.show()
```
