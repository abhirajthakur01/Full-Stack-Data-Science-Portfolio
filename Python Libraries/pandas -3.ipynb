```python
df.drop(columns=["New_Price"],inplace = True)
```

```python
#spliting the mileage column into mileage_value and mileage unit
df[['Mileage_value', 'Mileage_unit']] = df['Mileage'].str.extract(r'(\d+\.?\d*)\s*(.*)', expand=True)
```

```python
df.head()
```

```python
df[['Engine_value', 'Engine_unit']] = df['Engine'].str.extract(r'(\d+\.?\d*)\s*(.*)', expand=True)
```

```python
df.head()
```

```python
df[['power_value', 'power_unit']] = df['Power'].str.extract(r'(\d+\.?\d*)\s*(.*)', expand=True)
```

```python
df.head()
```

```python
df.drop(columns=["Power","Engine","Mileage"],inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
#converting mileage engine power value into numeric
df['Mileage_value']=df['Mileage_value'].astype('float')
```

```python
df['Engine_value']=df['Engine_value'].astype('float')
```

```python
df["power_value"]=pd.to_numeric(df["power_value"],errors="coerce")
```

```python
meidan_col = ["Kilometers_Driven"]
cat_col=["Location","Year","Transmission","Owner_Type","Colour","Seats","No. of Doors"]
```

```python
df.isnull().sum()
```

```python
for i in df.columns:
    if i in meidan_col:
        df[i].fillna(df[i].median(),inplace =True)
    elif i in cat_col:
        df[i].fillna(df[i].mode()[0],inplace=True)
    else:
        continue
```

```python
unit = ["Engine_unit","power_unit"]
median = ["power_value","Engine_value"]
```

```python
for i in df.columns:
    if i in unit:
        df[i].fillna(df[i].mode()[0],inplace=True)
    elif i in median:
        df[i].fillna(df[i].median(),inplace=True)
    else:
        continue
```
