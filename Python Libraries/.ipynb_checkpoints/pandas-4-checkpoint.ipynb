```python
#Dealing with null values
df.loc[:,"App Name":"Price"]
```

```python
#droping the develop website and privacy policy columns because there are 20% value are null
df.drop(columns = ["Developer Website","Privacy Policy"],inplace = True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
Mod_col = ["Category","Installs","Free","Currency","Minimum Android","Content Rating ","Ad Supported","In App Purchases","Editors Choice " ]
```

```python
for i in df.columns:
    if i in Mod_col:
        df[i].fillna(df[i].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
Mean_col= ["Rating","Rating Count","Minimum Installs","Maximum Installs", "Price"]
```

```python
for i in df.columns:
    if i in Mean_col:
        sns.boxplot(df[i])
        plt.show()
```

```python
for i in df.columns:
    if i in Mean_col:
        df[i].fillna(df[i].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df["Released"]= pd.to_datetime(df["Released"])
```

```python
df["Released_Year"] = df["Released"].dt.year
df["Released_Month"] = df["Released"].dt.month
df["Released_Month_Name"] = df["Released"].dt.month_name()
df["Released_Day_Name"] = df["Released"].dt.day_name()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset = ["Size"],inplace = True)
```

```python
df.isnull().sum()
```

```python
#Filling all the null values by the mode now
for i in df.columns:
    df[i].fillna(df[i].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```
