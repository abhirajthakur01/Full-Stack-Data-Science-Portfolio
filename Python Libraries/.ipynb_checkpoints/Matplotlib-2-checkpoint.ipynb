```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("churn_cleaned.xls")
```

```python
df.head()
```

```python
df.drop(columns=["Unnamed: 0"],inplace = True)
```

```python
df.head()
```

```python
index = df["Geography"].value_counts().index
value = df["Geography"].value_counts().values
```

```python
plt.bar(index,value,color = "g")
plt.xlabel("Geography")
plt.ylabel("Population")
plt.title("Geography wise population",fontsize = 15,color = "r")
plt.show()
```

```python
df.columns
```

```python
plt.bar(df['Gender'],df['EstimatedSalary'].mean())
plt.show()
```

```python
x = df.groupby("Gender")["EstimatedSalary"].mean().index
y = df.groupby("Gender")["EstimatedSalary"].mean().values
```

```python
plt.bar(x,y)
plt.show()
```

```python
x = df.groupby("Gender")["EstimatedSalary"].sum().index
y = df.groupby("Gender")["EstimatedSalary"].sum().values
```

```python
plt.barh(x,y)
plt.show()
```
