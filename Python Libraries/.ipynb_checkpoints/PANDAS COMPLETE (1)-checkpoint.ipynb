If we want to do operation on the original series we can use a parameter called inplace

directly performed on the original dataset

series

inplace = True  --> operation performed directly on series

```python
import pandas as pd 
```

```python
1) Data Frame is a pandas data structure
2 Data Frame is a 2 Dimensional labelled array
3 Rows in the Data Frame are heterogenous whereas
the columns are homogenous
```

```python
gh
```

If we want to get these rows where Height >160
here we use loc as it is labelled as Height
instead of 0,1 2,3 default values
here we access rows based on the condition

**iloc cannot be used when we are accessing
based on condition**

**how to access rows based on a single column
condition**

```python
gh.loc[gh['height']>160]
```

when we want to access rows based on
multiple condition

we use paranthesis between conditions with
& (logical_and) and | (logical-or) in between
them

we want people of Height >160 and male
candidate

```python
gh.loc[(gh['height']>160) & (gh['Gender']=='M')]
```

If we want to get only one column based on
these conditions

```python
d4 = gh.loc[(gh['height']>150)| (gh['Gender']=='M'),['name']]
```

```python
d4
```

**If we want to just see name & Age**

```python
d5 = gh.loc[(gh['height']>150)| (gh['Gender']=='M'),['name','age']]
```

```python
d5
```
