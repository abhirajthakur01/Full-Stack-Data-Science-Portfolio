```python
print("x =", x)
print("x + 5 =", x + 5)
print("x - 5 =", x - 5)
print("x * 2 =", x * 2)
print("x / 2 =", x / 2)
print("x // 2 =", x // 2) # floor division  
print("-x = ", -x)
print("x  2 = ",x ,2)
print("x % 2 = ",x % 2)
```
