#### bi-variate analysis

```python
x = df.groupby("Year")["Kilometers_Driven"].mean().sort_values().index
y = df.groupby("Year")["Kilometers_Driven"].mean().sort_values().values
```

```python
plt.figure(figsize = (10,10))
plt.style.use("dark_background")
sns.lineplot(x = x,y=y,color = "b")
plt.xticks(x,rotation = 90)
plt.title("kilometers  driven by number of year")
plt.ylabel("Kilometers driven")
plt.xlabel("Year")
plt.show()
```

```python
x = df.groupby("Location")["Kilometers_Driven"].mean().index
y = df.groupby("Location")["Kilometers_Driven"].mean().values
```

```python
plt.figure(figsize = (10,10))
plt.style.use("dark_background")
sns.lineplot(x = x,y=y,color = "b")
plt.xticks(x,rotation = 90)
plt.title("kilometers  driven by location")
plt.ylabel("Kilometers driven")
plt.xlabel("location")
plt.show()
```

```python
x = df.groupby("Location")["Price"].mean().index
y = df.groupby("Location")["Price"].mean().values
```

```python
plt.figure(figsize = (10,5))
plt.style.use("dark_background")
sns.lineplot(x = x,y=y,color = "b",marker = "*",mfc = "r", ms = 10)
plt.xticks(x,rotation = 90)
plt.title("price per location")
plt.ylabel("Kilometers driven")
plt.xlabel("price")
plt.show()
```
